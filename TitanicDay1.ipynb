```python
df = pd.read_csv("/content/train.csv")
```

Splitting the Data into Numerical and Categorical Data

```python
df_num = df[['Age', 'SibSp', 'Parch', 'Fare']]
df_cat = df[['Survived', 'Pclass', 'Sex', 'Ticket', 'Cabin', 'Embarked']]
```

```python
for i in df_num:
  plt.hist(df_num[i])
  plt.title(i)
  plt.show()
```

Now we want to take a look at the relationship between the numerical variables, with the .corr() function

```python
print(df_num.corr())
```

```python
sns.heatmap(df_num.corr())
```

```python
pd.pivot_table(df, index = 'Survived', values = ['Age', 'SibSp', 'Parch', 'Fare'])
```

```python
for i in df_cat:
    sns.countplot(x=df_cat[i])  # Plot counts of each category
    plt.title(i)  # Title will show the name of the categorical variable
    plt.show()
```

As we can see, a Countplot is not a good way to show some of the categorical data because some of them are unique, such as plane ticket.
